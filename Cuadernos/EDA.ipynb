```python
import pandas as pd 
```

```python
df_riesgo_economico = pd.read_csv("/Users/rodri/Desktop/Proyecto final/Datos/df_riesgo_economico_social_limpio.csv")
df_riesgo_economico
```

```python
df_riesgo_economico["Año"]=df_riesgo_economico["Año"].astype(str)
```

```python
df_riesgo_economico["Indice de Gini"]=df_riesgo_economico["Indice de Gini"].str.replace(',', '.').astype(float)
```

```python
df_riesgo_economico.describe()
```

```python
import matplotlib.pyplot as plt

#Analisis grafico de indicadores economicos 
variables = ['Crecimiento_GDP(%)', 'Inflacion Anual(%)', 'Balanza_comercial', 'Reservas Internacionales', 'Saldo Cuenta Corriente', 'Tipo de cambio Nominal', 'Deficit Publico (%PIB)']

#histogramas para cada variable
for var in variables:
    plt.figure(figsize=(8, 5))
    plt.hist(df_riesgo_economico[var].dropna(), bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(df_riesgo_economico[var].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Media: {df_riesgo_economico[var].mean():.2f}')
    plt.axvline(df_riesgo_economico[var].median(), color='green', linestyle='dashed', linewidth=1, label=f'Mediana: {df_riesgo_economico[var].median():.2f}')
    plt.title(f'Distribución de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.show()
```

```python
#Analisis grafico de indicadores sociales
variables = ['IDH', 'Riesgo Pobreza (%)', 'Indice de Gini', 'Tasa de desempleo']

#histogramas para cada variable
for var in variables:
    plt.figure(figsize=(8, 5))
    plt.hist(df_riesgo_economico[var].dropna(), bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(df_riesgo_economico[var].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Media: {df_riesgo_economico[var].mean():.2f}')
    plt.axvline(df_riesgo_economico[var].median(), color='green', linestyle='dashed', linewidth=1, label=f'Mediana: {df_riesgo_economico[var].median():.2f}')
    plt.title(f'Distribución de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.show()
```

```python
# Indicadores económicos
indicadores = ['Crecimiento_GDP(%)', 'Inflacion Anual(%)', 'Saldo Cuenta Corriente', 
               'Reservas Internacionales', 'Tipo de cambio Nominal', 'Deficit Publico (%PIB)', 
               'Balanza_comercial']

# Iterar sobre cada indicador para crear gráficos de líneas
for indicador in indicadores:
    plt.figure(figsize=(10, 6))
    
    # Graficar para cada país
    for pais in df_riesgo_economico['Country Name'].unique():
        data_pais = df_riesgo_economico[df_riesgo_economico['Country Name'] == pais]
        plt.plot(data_pais['Año'], data_pais[indicador], marker='o', label=pais)
    
    plt.title(f'Fluctuación del indicador: {indicador} a lo largo de los años')
    plt.xlabel('Año')
    plt.ylabel(indicador)
    
    # Colocar la leyenda fuera del área del gráfico
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Seleccionar solo las columnas numéricas del dataset
indicadores = ['Crecimiento_GDP(%)', 'Inflacion Anual(%)', 'Saldo Cuenta Corriente', 
               'Reservas Internacionales', 'Tipo de cambio Nominal', 'Deficit Publico (%PIB)', 
               'Balanza_comercial', 'IDH', 'Riesgo Pobreza (%)', 'Indice de Gini', 'Tasa de desempleo', 'IED']

# Crear un subset del DataFrame solo con las columnas relevantes
df_correlacion = df_riesgo_economico[indicadores]

# Convertir columnas numéricas que puedan contener comas a puntos
df_correlacion = df_correlacion.apply(lambda x: x.str.replace(',', '.').astype(float) if x.dtype == 'object' else x)

# Calcular la matriz de correlación de todas las columnas con la IED
correlacion_ied = df_correlacion.corr()[['IED']].sort_values(by='IED', ascending=False)
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))

sns.heatmap(correlacion_ied, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Mapa de Correlación de Indicadores con IED')
plt.show()
```
