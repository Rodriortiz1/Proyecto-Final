```python
from prophet import Prophet
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score

def evalue_prediction(df_riesgo_economico_social, name):
    # Preparar los datos para Prophet
    df_prophet = df_riesgo_economico_social[['Año', 'IED']].rename(columns={'Año': 'ds', 'IED': 'y'})
    # Convertir el año en un formato de fecha
    df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')

    # Separar los datos de entrenamiento (hasta 2021)
    df_train_prophet = df_prophet[df_prophet['ds'].dt.year <= 2021]

    # Separar los datos de prueba (año 2022)
    df_test_prophet = df_prophet[df_prophet['ds'].dt.year == 2022]
    # Crear y ajustar el modelo Prophet
    model_prophet = Prophet()
    model_prophet.fit(df_train_prophet)
    # Crear un DataFrame con las fechas futuras para la predicción
    future = model_prophet.make_future_dataframe(periods=2, freq='Y')  # 1 año en el futuro (2022)

    # Realizar la predicción
    forecast = model_prophet.predict(future)

    # Extraer la predicción para 2022
    forecast_2022 = forecast[forecast['ds'].dt.year == 2022]
    return df_test_prophet['y'], forecast_2022['yhat']
```

```python
df_final = pd.DataFrame(columns=['Country_name','2022','2023'])
for city in df_riesgo_economico_social['Country Name'].unique():
    df = df_riesgo_economico_social[df_riesgo_economico_social['Country Name'] == city]
    df_test_prophet , forecast_2022  = evalue_prediction(df, name=city)

    df_final.loc[len(df_final.index)] = [city, str(df_test_prophet).split()[1] , str(forecast_2022).split()[1]]
```
