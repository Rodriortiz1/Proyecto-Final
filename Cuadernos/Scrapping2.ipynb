```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página web que quieres scrapear
url = "https://datosmacro.expansion.com/comercio/balanza"

# Realiza la solicitud a la página web
response = requests.get(url)

# Asegúrate de que la solicitud fue exitosa
if response.status_code == 200:
    # Crea un objeto BeautifulSoup para analizar el HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Selecciona el elemento específico que quieres scrapear
    element = soup.select_one('body > div.dialog-off-canvas-main-canvas > div.main-container.container.js-quickedit-main-content > div > section > div > article > div > div:nth-child(4) > div > div.table-responsive')

    # Verifica si se encontró el elemento
    if element:
        # Encuentra las filas de la tabla dentro del elemento
        rows = element.find_all('tr')
        
        # Lista para almacenar los datos de la tabla
        table_data = []

        # Itera sobre las filas
        for row in rows:
            # Extrae los datos de cada columna
            cols = row.find_all('td')  # Si las celdas son 'td'
            cols = [col.text.strip() for col in cols]  # Limpia el texto en cada celda
            table_data.append(cols)
        
        # Convierte la lista de listas en un DataFrame de pandas
        df = pd.DataFrame(table_data)

        # Imprime el DataFrame
        print(df)
    else:
        print("Elemento no encontrado.")
else:
    print(f"Error al acceder a la página: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página
url = 'https://datosmacro.expansion.com/deficit?anio=2013'

# Realizar la solicitud HTTP
response = requests.get(url)
html_content = response.content

# Parsear el HTML con BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Seleccionar la tabla usando el selector CSS
table = soup.select_one('body > div.dialog-off-canvas-main-canvas > div.main-container.container.js-quickedit-main-content > div > section > div > article > div > div:nth-child(6) > div > div.table-responsive')

# Verificar si la tabla fue encontrada
if table:
    # Extraer datos de la tabla
    rows = table.find_all('tr')
    data = []
    for row in rows:
        cols = row.find_all('td')
        data.append([col.text.strip() for col in cols])
    
    # Convertir a DataFrame y guardar en CSV
    df = pd.DataFrame(data)
    df.to_csv('deficit_data.csv', index=False, header=False)
    print("Datos extraídos y guardados en deficit_data.csv")
else:
    print("No se encontró la tabla en el selector proporcionado.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página que quieres scrapear
url = "https://www.eustat.eus/elementos/ele0013500/ti_indice-de-desarrollo-humano-por-indicadores-segun-paises-2019/tbl0013566_c.html"

# Hacer la solicitud HTTP
response = requests.get(url)

# Comprobar que la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Seleccionar el elemento usando el selector CSS proporcionado
    container = soup.select_one("#containerID > div.col-lg-12.col-md-12.col-sm-12.col-xs-12 > div")
    
    if container:
        # Asumiendo que el contenedor contiene una tabla HTML
        table = container.find("table")
        
        if table:
            # Leer la tabla en un DataFrame de pandas
            df = pd.read_html(str(table))[0]  # Convertir la tabla a DataFrame
            
            # Guardar el DataFrame en un archivo CSV
            df.to_csv("indice_desarrollo_humano_2019.csv", index=False)
            
            print("Datos guardados en 'indice_desarrollo_humano_2019.csv'")
        else:
            print("No se encontró una tabla en el contenedor.")
    else:
        print("No se encontró el contenedor con el selector proporcionado.")
else:
    print(f"Error al hacer la solicitud: {response.status_code}")
```
