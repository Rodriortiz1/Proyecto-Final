## INDICE DE DESARROLLO HUMANO

```python
import pandas as pd
df_idh_2013 = pd.read_csv("/Users/rodri/IDH_2013.csv")
df_idh_2014 = pd.read_csv("/Users/rodri/IDH_2014.csv")
df_idh_2015 = pd.read_csv("/Users/rodri/IDH_2015.csv")
df_idh_2016 = pd.read_csv("/Users/rodri/IDH_2016.csv")
df_idh_2017 = pd.read_csv("/Users/rodri/IDH_2017.csv")
df_idh_2018 = pd.read_csv("/Users/rodri/IDH_2018.csv")
df_idh_2019 = pd.read_csv("/Users/rodri/IDH_2019.csv")
df_idh_2020 = pd.read_csv("/Users/rodri/IDH_2020.csv")
df_idh_2021 = pd.read_csv("/Users/rodri/IDH_2021.csv")
```

```python
def limpiar_idh (df):
    df.rename(columns= {"Unnamed: 0": "Country Name"}, inplace= True)
    df.rename(columns= {"Unnamed: 1": "IDH"}, inplace= True)
    df.drop(columns= ["Unnamed: 2","Unnamed: 3","Unnamed: 4"],inplace=True)
    df["Country Name"]=df["Country Name"].str.strip(" [+]")
    return df
```

```python
limpiar_idh(df_idh_2013)
limpiar_idh(df_idh_2014)
limpiar_idh(df_idh_2015)
limpiar_idh(df_idh_2016)
limpiar_idh(df_idh_2017)
limpiar_idh(df_idh_2018)
limpiar_idh(df_idh_2019)
limpiar_idh(df_idh_2020)
limpiar_idh(df_idh_2021)
```

```python
df_idh_2013["Año"]=2013
df_idh_2014["Año"]=2014
df_idh_2015["Año"]=2015
df_idh_2016["Año"]=2016
df_idh_2017["Año"]=2017
df_idh_2018["Año"]=2018
df_idh_2019["Año"]=2019
df_idh_2020["Año"]=2020
df_idh_2021["Año"]=2021
```

```python
dataframes= [df_idh_2013, df_idh_2014, df_idh_2015, df_idh_2016, df_idh_2017, df_idh_2018, df_idh_2019, df_idh_2020, df_idh_2021]
df_idh = pd.concat(dataframes, ignore_index=True)
df_idh 
```

```python
df_idh_2022 = pd.read_csv("/Users/rodri/IDH_2022_REAL.csv")
```

```python
df_idh_2022.drop(columns=["Unnamed: 0_level_0","Esperanza de vida al nacer en años","Años esperados de escolaridad (a)","Años promedio de escolaridad (a)","Renta nacional bruta per cápita en PPP en US$ de 2017 (*)"],inplace=True)
df_idh_2022.rename(columns={"Unnamed: 1_level_0":"Country Name"},inplace=True)
df_idh_2022.drop([0],axis=0,inplace=True)
```

```python
no_paises = ['GRUPOS DE DESARROLLO HUMANO',
       'DESARROLLO HUMANO MUY ALTO (0,800 y más)',
       'DESARROLLO HUMANO ALTO (0,700-0,799)',
       'DESARROLLO HUMANO MEDIO (0,550-0,699)',
       'DESARROLLO HUMANO BAJO (menos de 0,550)', 'TOTAL MUNDIAL','DESARROLLO HUMANO BAJO','DESARROLLO HUMANO MEDIO','DESARROLLO HUMANO MUY ALTO','DESARROLLO HUMANO ALTO','C.A. de Euskadi']

df_idh_2022 = df_idh_2022[~df_idh_2022['Country Name'].isin(no_paises)]
df_idh_2022["Country Name"].unique()
```

```python
df_idh_2022.dropna(inplace=True)
df_idh_2022["Año"]=2022
```

```python
df_idh_2022.reset_index(drop=True,inplace=True)
```

```python
df_idh_2022.rename(columns={"Índice de desarrollo humano": "IDH"},inplace=True)
df_idh_2022['IDH'] = df_idh_2022['IDH'].astype(float)
df_idh_2022['IDH'] = df_idh_2022['IDH'] / 1000
```

```python
df_idh = pd.concat([df_idh,df_idh_2022],ignore_index= True)
```

```python
df_idh.isnull().sum()
```

```python
df_idh.duplicated().sum()
```

```python
df_idh.to_csv("df_idh_limpio.csv",index=False)
```
